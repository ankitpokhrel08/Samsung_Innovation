```python
import folium

# Create a map centered between Portugal and China
m = folium.Map(location=[40, 60], zoom_start=4)  

# Add marker for the starting point in Portugal
folium.Marker(
    location=[37.0, -8.9],  # Near Sagres, Portugal
    popup="Sagres, Portugal",
).add_to(m)

# Add marker for the ending point in China
folium.Marker(
    location=[30.0, 120.0],  # Near Jiangsu, China
    popup="Jiangsu, China",
).add_to(m)

# Draw a straight line between Portugal and China
folium.PolyLine(
    locations=[[37.0, -8.9], [30.0, 120.0]],
    color="red",
    weight=5,
    opacity=0.7
).add_to(m)

# Display the map
m
```

```python
import folium

# Initialize the map centered around the midpoint of the route
m = folium.Map(location=[27.8, 84.0], zoom_start=7)

# List of cities with their coordinates and popups
cities = [
    {"name": "Kathmandu", "location": [27.7103, 85.3222]},
    {"name": "Hetauda", "location": [27.4287, 85.0322]},
    {"name": "Bharatpur", "location": [27.6766, 84.4357]},
    {"name": "Butwal", "location": [27.7000, 83.4500]},
    {"name": "Ghorahi", "location": [28.0419, 82.4853]},
    {"name": "Pyuthan", "location": [28.1017, 82.8533]},
]

# Add markers for each city
for city in cities:
    folium.Marker(
        location=city["location"],
        popup=city["name"],
    ).add_to(m)

# Extract the locations for the PolyLine
route = [city["location"] for city in cities]

# Draw the route connecting the cities
folium.PolyLine(
    locations=route,
    color="blue",
    weight=5,
    opacity=0.7
).add_to(m)

# Display the map
m
```
